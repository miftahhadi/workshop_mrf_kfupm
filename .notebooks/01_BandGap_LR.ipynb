## **Siapkan Data**

Pisahkan dulu mana features dan mana output.

Karena di linear regression kita cuma bisa menghitung data numerik, jadi tipe ```string``` (dalam kasus kita: kolom ```composition```), perlu dihilangkan dulu.

````python
# Separate features and target variable
X = data.drop(columns=['composition','gap expt'])
y = data['gap expt']
````

````python
X.head()
````

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

Pisahkan data jadi 2: **training set** & **test set**

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42
)
````

## **Saatnya Buat Model ML** ðŸ”¥

Kita mau pakai ```LinearRegression```. Panggil dulu dari Scikit-Learn

````python
# Import dulu modulnya
from sklearn.linear_model import LinearRegression

# Buat modelnya maujud
model = LinearRegression()
````

Ayo kita fitting ke data training

````python
model.fit(X_train, y_train)
````

Sekarang coba prediksi data test

````python
y_predict = model.predict(X_test)
````

## **Seberapa Bagus Model Kita Ini?**

Import modul untuk scoring

````python
from sklearn.metrics import mean_squared_error, root_mean_squared_error,r2_score
````

````python
r2 = r2_score(y_test, y_predict)
print("Skor R^2: ", r2)

mse = mean_squared_error(y_test, y_predict)
print("Skor MSE: ", mse)

rmse = root_mean_squared_error(y_test, y_predict)
print("Skor RMSE: ", rmse)
````

````python
data['predicted_gap'] = model.predict(X_scaled)
data['percentage_error'] = abs((data['predicted_gap'] - data['gap expt']) / data['gap expt']) * 100
````

````python
import plotly.express as px
import plotly.graph_objects as go 

fig = px.scatter(
    data,
    x="predicted_gap",
    y="gap expt",
    hover_name="composition",
    color="percentage_error",
    title="Predicted vs Experimental Band Gaps",
    color_continuous_scale=px.colors.sequential.Bluered,
    width=1200,
    height=800
)

reference_line = go.Scatter(
    x=[0, 12],
    y=[0,12] ,
    line=dict(color='red', dash='dash'),
    mode="lines",
    showlegend=False
)

fig.add_trace(reference_line)
fig.show()
````

## **Cross-Validation**

````python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])
````

````python
from sklearn.model_selection import KFold, cross_validate

kfold = KFold(n_splits=5, shuffle=True)
````

````python
scoring = ["r2", "neg_mean_squared_error"]

cv_results = cross_validate(
    pipeline,
    X_train, y_train,
    cv=kfold,
    scoring=scoring,
    return_train_score=True
)
````

````python
r2_test = cv_results["test_r2"]
mse_test = -cv_results["test_neg_mean_squared_error"]
rmse_test = np.sqrt(mse_test)

print("RÂ² per fold:", r2_test)
print("Mean RÂ²:", r2_test.mean())
print("Std RÂ²:", r2_test.std())

print("MSE per fold:", mse_test)
print("Mean MSE:", mse_test.mean())

print("RMSE per fold:", rmse_test)
print("Mean RMSE:", rmse_test.mean())
````

````python
# >>> Final training on full training set <<<
pipeline.fit(X_train, y_train)  
````

````python
y_pred = pipeline.predict(X_test)

print("Test RÂ²:", r2_score(y_test, y_pred))
print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test RMSE:", root_mean_squared_error(y_test, y_pred))
````
